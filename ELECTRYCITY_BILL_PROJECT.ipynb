```python
users=[]
```

```python
def calculate_bill(units):
    if 1 <=units<=100:
        return units * 10
    elif 101<=units<=200:
        return units * 15
    elif 201<=units<=300:
        return units * 20
    else:
        return units * 25
    #units = result
    
```

```python
# 1:
def add_user():
    print('===== ADD NEW USER =====')
    name= input("Enter User Name : ")
    cnic = input("Enter your cnic (xxxx-xxxxxxx-x)")
    if len(cnic)== 15 and cnic[5] == "-" and cnic[13] == '-':
         cnic
    else:
        print("invalid number! pick up the pattrn.")
        
    for user in users:
        if user['cnic'] == cnic:
            print("user already exist")
            return
    try:
        units=int(input("Enter useable unit :"))
    except ValueError:
                  print("invalid number! Units must be integer.")
    bill=calculate_bill(units)
    user={
        'name' : name,
        'cnic' : cnic,
        'units': units,
        'bill' : bill
    }
    users.append(user)
    print("\n User Successfully Added.")
        
    
```

```python
# show all user 2
def show_all_users():
    
    if not users:
        print(" User not found")
        return
    print("=== Show All User Data ===") 
    print(f"\n{'name':<15} {'cnic':<20} {'units':<15} {'bill':<4}")
    print('-'*70)
    for i,user in enumerate(users , 1):
       # print(f"\nUser {i}:")
        print(f"{i} : {user['name']:<15} {user['cnic']:<20} {user['units']:<15} {user['bill']:<4}PKR")         
       # print("Name of User : ", user['name'])
       # print("Cnic of User :", user['cnic'])
        #print("unit useable :", user['units'])
        #print("Total Bill :", user['bill'])C
    
          
```

```python
# 3
def search_user():
    
    cnic = input("Enter search cnic :")
    found = False
    for user in users:
           if user['cnic'] == cnic:
               print("=== User Found === ")
               print("Name of User : ", user['name'])
               print("Cnic of User :", user['cnic'])
               print("unit useable :", user['units'])
               print("Total Bill :", user['bill'])
               found = True
               break
                   
    if not found:
        print("\nUser not found")                        
```

```python
# 4
def delete_user():
    if not users:
        print("user not found")
        return
    cnic = input("Enter delete cnic :")
    for user in users:
        if user['cnic'] == cnic:
            users.remove(user)
            print(f"User {user['name']} with CNIC {user['cnic']} delete successfully.")
            break
```

```python
# 5
def show_total_revenue():
    print("\n ===== Total Revenue =====")
    total = sum(user["bill"] for user in users)
   # total = sum(user["bill"] for user in users)
    print(f"\n Total Revenue from All Users: {total} PKR")
```
